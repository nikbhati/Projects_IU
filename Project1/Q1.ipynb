```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import Lasso,ElasticNet,Ridge,LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# as there are no empty/null values we don't need to use Simple Imputor 
```

```python
data.columns
```

```python
ct= ColumnTransformer([
    ("scaling",StandardScaler(),['sqft_living','sqft_lot','sqft_above','sqft_basement']),
    ("onehot", OneHotEncoder(sparse=False,handle_unknown='ignore'),['bedrooms','bathrooms','floors','waterfront','view','condition'])])
```

```python
ct1=ColumnTransformer([("onehot", OneHotEncoder(sparse=False,handle_unknown='ignore'),['bedrooms','bathrooms','floors','waterfront','view','condition'])
])
```

```python
X=data.drop('price',axis=1)
y=data.price
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0)
```

```python
ct.fit(X_train)
```

```python
X_train_trans=ct.transform(X_train)
X_test_trans=ct.transform(X_test)
```

```python
from sklearn.linear_model import Lasso,ElasticNet,Ridge,LinearRegression
```

```python
lr=LinearRegression()
ridge=Ridge()
lasso=Lasso()
en=ElasticNet()
```

```python
algo=[lr,lasso,ridge,en]
```

```python
for i in algo:
    i.fit(X_train_trans,y_train)
    acc=i.score(X_test_trans,y_test)
    print(acc,i)
```

```python
ct1.fit(X_train)
```

```python
X_train_trans2=ct1.transform(X_train)
X_test_trans2=ct1.transform(X_test)
```

```python
lr=LinearRegression()
ridge=Ridge()
lasso=Lasso()
en=ElasticNet()
```

```python
algo=[lr,lasso,ridge,en]

for i in algo:
    i.fit(X_train_trans2,y_train)
    acc=i.score(X_test_trans2,y_test)
    print(acc,i)
```

```python
# thus we can conclude that scaling improves the accuracy drastically 
# the above accuracies are calculated without scaling and the ones with scaling have larger values
```
